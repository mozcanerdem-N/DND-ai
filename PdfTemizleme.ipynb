**2. Pdf dosyasını temize geçirmek** DENEME 1

```python
import fitz  # PyMuPDF
import re
from google.colab import files

# PDF yükleme
uploaded = files.upload()

pdf_path = list(uploaded.keys())[0]

# PDF açma
doc = fitz.open(pdf_path)

text = ""

# Sayfalardaki metni alma
for page in doc:
    text += page.get_text()

    # ---- TEMİZLEME ----

    # satır sonu bölünmüş kelimeleri birleştir
    text = re.sub(r'-\n', '', text)

    # gereksiz satır sonlarını boşluk yap
    text = re.sub(r'\n+', '\n', text)

    # fazla boşlukları temizle
    text = re.sub(r'[ \t]+', ' ', text)

    # baştaki ve sondaki boşlukları sil
    text = text.strip()

    # ---- TXT DOSYASI OLUŞTUR ----
    txt_file = "temiz_metin.txt"

    with open(txt_file, "w", encoding="utf-8") as f:
        f.write(text)

        # indir
        files.download(txt_file)

        print("Temizleme tamamlandı.")
```

**3. Pdf dosyasını temize geçirmek** DENEME 2

```python
!pip install pymupdf

import fitz
import re
from datetime import datetime, timezone, timedelta
from google.colab import files

# PDF yükle
uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]

doc = fitz.open(pdf_path)

text = ""
for page in doc:
    text += page.get_text()

# -------- TEMİZLEME --------

# bölünmüş kelimeler
text = re.sub(r'-\n', '', text)

# satır başı boşlukları kaldır
text = re.sub(r'^[ \t]+', '', text, flags=re.MULTILINE)

# fazla boşlukları temizle
text = re.sub(r'[ \t]+', ' ', text)

# fazla boş satırları azalt
text = re.sub(r'\n{2,}', '\n\n', text)

# noktalama öncesi boşlukları kaldır (TDK)
text = re.sub(r'\s+([.,;:!?])', r'\1', text)

# noktalama sonrası tek boşluk bırak
text = re.sub(r'([.,;:!?])([^\s])', r'\1 \2', text)

# parantez boşlukları
text = re.sub(r'\(\s+', '(', text)
text = re.sub(r'\s+\)', ')', text)

text = text.strip()

# -------- DOSYA --------

utc_minus8 = timezone(timedelta(hours=-8))
timestamp = datetime.now(utc_minus8).strftime("%Y-%m-%d_%H-%M")

filename = f"yonetmelik_temiz_{timestamp}.txt"

with open(filename, "w", encoding="utf-8") as f:
    f.write(text)

files.download(filename)

print("Temizleme tamamlandı:", filename)
```

```python
!pip install pymupdf

import fitz
import re
from datetime import datetime, timezone, timedelta
from google.colab import files

uploaded = files.upload()
pdf_path = list(uploaded.keys())[0]

doc = fitz.open(pdf_path)

text = ""
for page in doc:
    text += page.get_text()

# bölünmüş kelimeler
text = re.sub(r'-\n', '', text)

# satır başı boşlukları kaldır
text = re.sub(r'^[ \t]+', '', text, flags=re.MULTILINE)

# fazla boşlukları temizle
text = re.sub(r'[ \t]+', ' ', text)

# BOŞ SATIRLARI KALDIR
text = re.sub(r'\n\s*\n', '\n', text)

# TDK noktalama
text = re.sub(r'\s+([.,;:!?])', r'\1', text)
text = re.sub(r'([.,;:!?])([^\s])', r'\1 \2', text)

text = text.strip()

# dosya oluştur
utc_minus8 = timezone(timedelta(hours=-8))
timestamp = datetime.now(utc_minus8).strftime("%Y-%m-%d_%H-%M")

filename = f"temiz_metin_{timestamp}.txt"

with open(filename, "w", encoding="utf-8") as f:
    f.write(text)

files.download(filename)

print("Tamamlandı:", filename)
```
